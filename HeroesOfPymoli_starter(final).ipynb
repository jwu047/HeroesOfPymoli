## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
print(purchase_data.head())

# Check for any special spellings within the column names
print(purchase_data.columns.unique())

# Basic calculations to obtain unique items, average prices, number of purchases
# and Total Revenue

# Number of Items using Item ID
unique_items = purchase_data["Item ID"].unique()
no_items = len(unique_items)
print("Number of items: " + str(no_items))

# Average Price, round up to the second decimal and put with the dollar symbol
average_price = purchase_data["Price"].mean()
average_price = round(average_price, 2)
average_price = '${}'.format(average_price)
print("Average Price: " + str(average_price))

# Number of purchases by counting the purchase id column
no_purchases = len(purchase_data["Purchase ID"])
print("Number of Purchases: " + str(no_purchases))

# Total Revenue
total_revenue = purchase_data["Price"].sum()
total_revenue = round(total_revenue, 2)
total_revenue = '${:,}'.format(total_revenue)
print("Total Revenue :" + str(total_revenue))
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
# Cut data into bins

# Run all cells for this to work
purchase_data["Age"] = pd.cut(purchase_data["Age"], age_bins, labels=(group_names))
#purchase_data["Age"]
purchase_data_group = purchase_data.groupby("Age")

age_df = purchase_data_group.count()
any_df = purchase_data_group["Price"].count()
#print(any_df)

# percentage of players, divide by total players, 576
a = round(age_df.loc["<10","Price"] / total_players * 100, 2)
b = round(age_df.loc["10-14","Price"] / total_players * 100, 2)
c = round(age_df.loc["15-19","Price"] / total_players * 100, 2)
d = round(age_df.loc["20-24","Price"] / total_players * 100, 2)
e = round(age_df.loc["25-29","Price"] / total_players * 100, 2)
f = round(age_df.loc["30-34","Price"] / total_players * 100, 2)
g = round(age_df.loc["35-39","Price"] / total_players * 100, 2)
h = round(age_df.loc["40+","Price"] / total_players * 100, 2)

arr = [a,b,c,d,e,f,g,h]
arr_df = pd.DataFrame(arr)
#print(arr_df)

# Ask for help on this whole cell, technically right to get the answer but inefficient
any_df.reset_index(drop=True, inplace=True)
arr_df.reset_index(drop=True, inplace=True)

final_df = pd.concat( [any_df, arr_df], axis=1)
final_df = final_df[[0,"Price"]]
final_df = final_df.rename(columns={0:"Percentage of Players", "Price": "Total Count"})
final_df.index = group_names
final_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_price = purchase_data.groupby("Item ID")["Price"].mean()
item_price = ['${:.2f}'.format(x) for x in item_price]
#len(item_price)
total_value = purchase_data.groupby("Item ID")["Price"].sum()
total_value = ['${:.2f}'.format(x) for x in total_value]
#len(total_value)
item_count = purchase_data.groupby("Item ID")["Price"].count()
#len(item_count)
item_name = purchase_data.groupby("Item ID")["Item Name"].unique()

popular_items_df = pd.DataFrame({"Item Name": item_name})
popular_items_df["Purchase Count"] = item_count
popular_items_df["Item Price"] = item_price
popular_items_df["Total Purchase Value"] = total_value

popular_items_df = popular_items_df.sort_values("Purchase Count", ascending = False)
popular_items_df.head()
```
